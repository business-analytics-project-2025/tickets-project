```python
import pandas as pd
```

```python
df = pd.read_csv("aa_dataset-tickets-multi-lang-5-2-50-version.csv")
```

```python
df = df[df["language"] != "de"]
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.drop(columns=["language", "version"])
```

```python
df["body"] = (
    df["body"]
      .str.replace(r"\\n\\n", " ", regex=True)
      .str.replace(r"\\n", " ", regex=True)
      .str.strip()
)
```

```python
df["answer"] = (
    df["answer"]
      .str.replace(r"\\n\\n", " ", regex=True)
      .str.replace(r"\\n", " ", regex=True)
      .str.strip()
)
```

```python
df["subject"] = (
    df["subject"]
      .str.replace(r"\\n\\n", " ", regex=True)
      .str.replace(r"\\n", " ", regex=True)
      .str.strip()
)
```

```python
df["answer"] = (
    df["answer"]
      .str.replace(r"\s*,?\s*<name>\s*,?\s*", " ", regex=True)
      .str.replace(r"\s+([.,!?;:])", r"\1", regex=True)
      .str.replace(r"\s{2,}", " ", regex=True)   # collapse double spaces
      .str.strip()
)
```

```python
df = df.drop(columns=["tag_4", "tag_5", "tag_6", "tag_7", "tag_8"])
```

```python
df = df.rename(columns={"queue":"department"})
```
